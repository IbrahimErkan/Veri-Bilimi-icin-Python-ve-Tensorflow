## Label

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y)
plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")
plt.show()
```

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y)
plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni", fontsize=25, color="red", loc="right", fontname="comic Sans Ms")
plt.ylabel("Y Ekseni", fontsize=25, color="red", loc="top", fontname="comic Sans Ms")
plt.show()
```

## Ticks

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y)
plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni", fontsize=25, color="red", loc="right", fontname="comic Sans Ms")
plt.ylabel("Y Ekseni", fontsize=25, color="red", loc="top", fontname="comic Sans Ms")

plt.xticks([1,2,3])
plt.show()
```

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y)
plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni", fontsize=25, color="red", loc="right", fontname="comic Sans Ms")
plt.ylabel("Y Ekseni", fontsize=25, color="red", loc="top", fontname="comic Sans Ms")

plt.xticks([1,2,3])
plt.yticks([2,6,10,14,18,22,26])
plt.show()
```

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y, label="label1")

plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni", fontsize=25, color="red", loc="right", fontname="comic Sans Ms")
plt.ylabel("Y Ekseni", fontsize=25, color="red", loc="top", fontname="comic Sans Ms")

plt.xticks([1,2,3])
plt.yticks([2,6,10,14,18,22,26])

plt.legend()
plt.show()
```

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y, label="label1", color="blue", linewidth=2.5, marker="o")

plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni", fontsize=25, color="red", loc="right", fontname="comic Sans Ms")
plt.ylabel("Y Ekseni", fontsize=25, color="red", loc="top", fontname="comic Sans Ms")

plt.xlim([-1,7])
plt.ylim([1,28])

plt.legend()
plt.show()
```

```python
x = [1,2,3]
y = [4,8,24]

plt.plot(x,y, label="label1", color="yellow", linewidth=2.5, marker="s", markersize=10, markeredgecolor="green")

plt.title("İlk Grafik", fontsize = 25, color = "blue",loc = "right", pad = 50, fontname="comic Sans Ms")

plt.xlabel("X Ekseni", fontsize=25, color="red", loc="right", fontname="comic Sans Ms")
plt.ylabel("Y Ekseni", fontsize=25, color="red", loc="top", fontname="comic Sans Ms")

plt.xlim([-1,7])
plt.ylim([1,28])

plt.legend()
plt.show()
```
